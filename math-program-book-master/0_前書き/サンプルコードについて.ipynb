```python
import pathlib

with open(str(pathlib.Path.cwd() / '..' / 'Pipfile')) as f:
    is_requires = False
    for r in f:
        r = r.strip()
        if is_requires:
            print('Python {}'.format(r.split('=')[1].replace('"','').strip()))
            break
        elif r == '[requires]':
            is_requires = True
```

```python
import pathlib

with open(str(pathlib.Path.cwd() / '..' / 'Pipfile')) as f:
    is_package_information = False
    for r in f:
        r = r.strip()
        if is_package_information:
            if r:
                r = r.split('"')[:2]
                r[0] = r[0].split('=')[0].strip()

                if r[1] == '*':
                    del r[1]

                print(''.join(r))
            else:
                break
        elif r == '[packages]':
            is_package_information = True
```
