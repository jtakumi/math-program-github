```python
%matplotlib inline
import math  # 標準のmathモジュール

import numpy as np  # NumPy
import pandas as pd  # Pandas

# m
from matplotlib import pylab as plt  # Matplotlib

# matplotlibで日本語表示する為の指定
import japanize_matplotlib
```

```python
# 東京, 仙台, 名古屋の座標（経度はcmなので1万を掛ける）
tokyo = [35.709 * 111, 139.732 * 91 * 10000]
nagoya = [35.181 * 111, 136.906 * 91 * 10000]
sendai = [38.254 * 111, 140.891 * 91 * 10000]
```

```python
from scipy.spatial.distance import euclidean

pd.DataFrame([[euclidean(tokyo, sendai)], [euclidean(tokyo, nagoya)]],
             index=['東京-仙台', '東京-名古屋'], columns=['ユーグリッド距離'])
```

```python
# 標準偏差を取っておく
std_x = np.std([tokyo[0], nagoya[0], sendai[0]])
std_y = np.std([tokyo[1], nagoya[1], sendai[1]])
std = [std_x, std_y]


def standard_euclidean(x, y, std):
    """
    標準ユークリッド距離
    :param x: x
    :param y: y
    :param std: 標準偏差
    :return: 標準ユークリッド距離
    """
    d = 0

    for i in range(len(x)):
        d += ((x[i] - y[i]) / std[i]) ** 2

    return np.sqrt(d)


pd.DataFrame([[standard_euclidean(tokyo, sendai, std)], [standard_euclidean(tokyo, nagoya, std)]],
             index=['東京-仙台', '東京-名古屋'], columns=['標準ユーグリッド距離'])
```
