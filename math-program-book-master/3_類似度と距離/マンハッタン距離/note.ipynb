```python
%matplotlib inline
import math  # 標準のmathモジュール

import numpy as np  # NumPy
import pandas as pd  # Pandas

# matplotlibで日本語表示する為の指定
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['MS PGothic', 'Osaka', 'Hiragino Sans', 'TakaoGothic', 'IPAexGothic']

from matplotlib import pylab as plt  # Matplotlib
```

```python
# 東京, 名古屋, 仙台の座標
tokyo = [35.709 * 111, 139.732 * 91]
nagoya = [35.181 * 111, 136.906 * 91]
sendai = [38.254 * 111, 140.891 * 91]
```

```python
def manhattan(x, y):
    """
    マンハッタン距離
    :param x: x
    :param y: y
    :return: マンハッタン距離
    """
    d = 0

    for i in range(len(x)):
        d += np.abs(x[i] - y[i])

    return d


pd.DataFrame([[manhattan(tokyo, sendai)], [manhattan(tokyo, nagoya)]],
             index=['東京-仙台', '東京-名古屋'], columns=['マンハッタン距離'])
```
