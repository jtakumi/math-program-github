```python
%matplotlib inline
import math  # 標準のmathモジュール

import numpy as np  # NumPy
import pandas as pd  # Pandas

from matplotlib import pylab as plt  # Matplotlib

# matplotlibで日本語表示する為の指定
import japanize_matplotlib
```

```python
# 1つめの要素が国語、2つ目の要素が数学
score = dict()
score['A'] = [55, 80]
score['B'] = [83, 58]
score['C'] = [70, 95]
score['D'] = [50, 40]
```

```python
fig, ax = plt.subplots()
plt.xlabel('国語 (点)')
plt.xlim([0, 100])
plt.ylabel('数学 (点)')
plt.ylim([0, 100])
for k, v in score.items():
    ax.annotate(k, xy=(v[0], v[1]))
    ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1)
```

```python
# 足し算
x = [1, 5]
y = [5, 3]
answer = [0, 0]

for i in range(len(x)):
    answer[i] = x[i] + y[i]

answer
```

```python
fig, ax = plt.subplots()
plt.xlim([0, 10])
plt.ylim([0, 10])
ax.annotate('x', xy=(x[0], x[1]))
ax.annotate('y', xy=(y[0], y[1]))
ax.annotate('answer', xy=(answer[0], answer[1]))
ax.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1)
ax.quiver(0, 0, y[0], y[1], angles='xy', scale_units='xy', scale=1)
ax.quiver(0, 0, answer[0], answer[1], angles='xy', scale_units='xy', scale=1)
```

```python
# 引き算
x = [1, 5]
y = [5, 3]
answer = [0, 0]

for i in range(len(x)):
    answer[i] = x[i] - y[i]

answer
```

```python
fig, ax = plt.subplots()
plt.xlim([-10, 10])
plt.ylim([-10, 10])
ax.annotate('x', xy=(x[0], x[1]))
ax.annotate('y', xy=(y[0], y[1]))
ax.annotate('-y', xy=(-y[0] - 0.5, -y[1]))
ax.annotate('answer', xy=(answer[0] - 2, answer[1]))
ax.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1)
ax.quiver(0, 0, y[0], y[1], angles='xy', scale_units='xy', scale=1)
ax.quiver(0, 0, -y[0], -y[1], angles='xy', scale_units='xy', scale=1)
ax.quiver(0, 0, answer[0], answer[1], angles='xy', scale_units='xy', scale=1)
```

```python
x = [1, 5]
answer = 0

for xt in x:
    answer += math.pow(xt, 2)

math.sqrt(answer)
```
