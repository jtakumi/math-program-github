```python
%matplotlib inline
import math  # 標準のmathモジュール

import numpy as np  # NumPy
import pandas as pd  # Pandas

from matplotlib import pylab as plt  # Matplotlib

# matplotlibで日本語表示する為の指定
import japanize_matplotlib
```

```python
# 東京, 名古屋, 仙台の座標
tokyo = [35.709 * 111, 139.732 * 91]
nagoya = [35.181 * 111, 136.906 * 91]
sendai = [38.254 * 111, 140.891 * 91]
```

```python
def euclidean(x, y):
    """
    ユーグリッド距離
    :param x: x座標
    :param y: y座標
    :return: ユーグリッド距離
    """
    d = 0

    for i in range(len(x)):
        d += (x[i] - y[i]) ** 2

    return np.sqrt(d)


pd.DataFrame([[euclidean(tokyo, sendai)], [euclidean(tokyo, nagoya)]],
             index=['東京-仙台', '東京-名古屋'], columns=['ユーグリッド距離'])
```

```python
from scipy.spatial.distance import euclidean

# 東京-仙台
euclidean(tokyo, sendai)
```

```python
tokyo = np.array(tokyo)
sendai = np.array(sendai)

# 東京-仙台
np.sqrt(np.sum((tokyo - sendai) ** 2))
```
