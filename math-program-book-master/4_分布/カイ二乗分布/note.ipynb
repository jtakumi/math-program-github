```python
%matplotlib inline
import math  # 標準のmathモジュール

import numpy as np  # NumPy
import pandas as pd  # Pandas

from matplotlib import pylab as plt  # Matplotlib

# matplotlibで日本語表示する為の指定
import japanize_matplotlib
```

```python
from scipy.special import gamma


def chi_squared_dist(x, k):
    """
    カイ二乗分布 (確率密度関数)
    :param x: x
    :param k: 自由度
    :return: カイ二乗分布 (確率密度関数)
    """
    numer = (1 / 2) ** (k / 2)
    denom = gamma(k / 2)
    other = x ** ((k / 2) - 1) * (math.e ** (-x / 2))
    return (numer / denom) * other


t = np.linspace(0.001, 4, 100)

for k in range(1, 5):
    y = [chi_squared_dist(x, k) for x in t]
    plt.plot(t, y, label='k=%d' % k)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.ylim([0.0, 1.0])
plt.title('グラフ1 : カイ二乗分布')
```

```python
from scipy.special import gamma


def chi_squared_dist(x, k):
    """
    カイ二乗分布 (確率密度関数)
    :param x: x
    :param k: 自由度
    :return: カイ二乗分布 (確率密度関数)
    """
    numer = (1 / 2) ** (k / 2)
    denom = gamma(k / 2)
    other = x ** ((k / 2) - 1) * (math.e ** (-x / 2))
    return (numer / denom) * other
```

```python
t = np.linspace(0.001, 80, 100)
k = 5
y = [chi_squared_dist(x, k) for x in t]
plt.plot(t, y, label='k=%d' % k)
```

```python
def chi_squared(e, x):
    """
    カイ二乗
    :param e: 期待値
    :param x: x
    :return: カイ二乗
    """
    result = 0

    for v in x:
        result += ((v - e) ** 2) / e

    return result


expect = 50 / 3
dice = [50, 10, 10, 10, 10, 10]
chi_squared(expect, dice)
```

```python
expect = 50 / 3
dice = [20, 20, 15, 15, 15, 15]
chi_squared(expect, dice)
```
