```python
%matplotlib inline
import math  # 標準のmathモジュール

import numpy as np  # NumPy
import pandas as pd  # Pandas

from matplotlib import pylab as plt  # Matplotlib

# matplotlibで日本語表示する為の指定
import japanize_matplotlib
```

```python
# 男性胸囲平均
male_mu = 86.5

# 男性胸囲標準偏差
male_sigma = 6.60

# 女性胸囲平均
female_mu = 82.6

# 女性胸囲標準偏差
female_sigma = 5.55
```

```python
from scipy.special import erf


def normal_dist_cdf(x, mu, sigma):
    """
    累積分布関数 (正規分布)
    :param x: x
    :param mu: 平均
    :param sigma: 標準偏差
    :return: 累積分布関数 (正規分布)
    """
    return 1 / 2 * (1 + erf((x - mu) / (sigma * math.sqrt(2))))


pd.DataFrame([[normal_dist_cdf(80, male_mu, male_sigma)],
              [normal_dist_cdf(80, female_mu, female_sigma)]], index=['男性', '女性'], columns=['胸囲が80以下である確率'])
```

```python
pd.DataFrame([[normal_dist_cdf(99, male_mu, male_sigma) - normal_dist_cdf(90, male_mu, male_sigma)],
              [normal_dist_cdf(99, female_mu, female_sigma) - normal_dist_cdf(90, female_mu, female_sigma)]],
             index=['男性', '女性'], columns=['胸囲が90〜99である確率'])
```

```python
# 解答例
from scipy.integrate import quad
from scipy.stats import norm

# 男性胸囲平均
male_mu = 86.5

# 男性胸囲標準偏差
male_sigma = 6.60

# 女性胸囲平均
female_mu = 82.6

# 女性胸囲標準偏差
female_sigma = 5.55

pd.DataFrame([[quad(norm(loc=male_mu, scale=male_sigma).pdf, 0, 80)[0]],
              [quad(norm(loc=female_mu, scale=female_sigma).pdf, 0, 80)[0]]],
             index=['男性', '女性'], columns=['胸囲が80以下である確率'])
```
