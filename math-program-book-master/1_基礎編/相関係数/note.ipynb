```python
%matplotlib inline
import math  # 標準のmathモジュール

import numpy as np  # NumPy
import pandas as pd  # Pandas

from matplotlib import pylab as plt  # Matplotlib

# matplotlibで日本語表示する為の指定
import japanize_matplotlib
```

```python
# プロ野球選手
baseball = pd.read_csv('../../9_data/プロ野球/プロ野球選手身長体重.csv')

# Jリーグ選手
soccer = pd.read_csv('../../9_data/サッカー/Jリーグ選手身長体重.csv')

# 力士
sumou = pd.read_csv('../../9_data/相撲/力士身長体重.csv')
```

```python
def corr(x, y):
    """
    相関係数を求める
    :param x: 標本x
    :param y: 標本y
    :return: 相関係数
    """
    return np.cov(x, y)[0][1] / (np.std(x, ddof=1) * np.std(y, ddof=1))


pd.DataFrame({'相関係数': [
    corr(baseball['身長'], baseball['体重']),
    corr(soccer['身長'], soccer['体重']),
    corr(sumou['身長'], sumou['体重'])
]}, index=['野球', 'サッカー', '相撲'])
```
