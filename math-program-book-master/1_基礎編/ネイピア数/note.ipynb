```python
%matplotlib inline
import math  # 標準のmathモジュール

import numpy as np  # NumPy
import pandas as pd  # Pandas

from matplotlib import pylab as plt  # Matplotlib

# matplotlibで日本語表示する為の指定
import japanize_matplotlib
```

```python
def e(n):
    """
    式a
    :param n: n
    :return: ネイピア数
    """
    return (1 + (1 / n)) ** n


# nに1〜100までを代入して式aの結果を取る
points = [e(n) for n in range(1, 100)]

# 結果のplot
plt.plot(points)
plt.title('グラフ1 : 自然対数の底')
plt.hlines([math.e], 1, 100, 'red', linestyles='dashed')  # ネイピア数を表す破線
```

```python
e(2 ** 32 - 1)
```

```python
# n=64で計算すると1.0になってしまう
# (1/ nが小さすぎて1を足すと情報落ちで1/ nの値がなかったことになるため発生する)
e(2 ** 64 - 1)
```

```python
from decimal import Decimal


def e(n):
    """
    式a
    :param n: n
    :return: ネイピア数
    """
    return (Decimal(1) + Decimal(1 / n)) ** n


e(2 ** 64 - 1)
```
