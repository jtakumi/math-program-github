```python
gdp = df['GDP実質成長率']
gdp = 1.0 + gdp / 100
pd.DataFrame(gdp.head())
```

```python
n = len(gdp)

# gmはgeometric meanの略
gm = 1.0

for i in range(n):
    gm *= gdp[i]

# 1/n乗する
gm **= (1 / n)

# 前年度比から成長率へ変形する
(gm - 1) * 100
```

```python
from scipy.stats import gmean

(gmean(gdp) - 1) * 100
```
