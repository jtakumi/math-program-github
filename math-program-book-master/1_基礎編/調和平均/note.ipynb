```python
def to_sec(t):
    """
    タイムを秒に直す
    :param t: タイム (hh:mm:ss)
    :return: 秒
    """
    ts = t.split(':')
    return int(ts[0]) * 3600 + int(ts[1]) * 60 + int(ts[2])


df = pd.read_csv('../../9_data/陸上/箱根駅伝2016青学.csv')

df['タイム (s)'] = df['タイム'].apply(to_sec)

# 各区間の平均時速を計算しておく
df['平均時速 (km/h)'] = 3600 * df['距離'] / df['タイム (s)']

df
```

```python
# 平均時速
3600 * df['距離'].sum() / df['タイム (s)'].sum()
```

```python
# 各区間の相加平均時速
df['平均時速 (km/h)'].mean()
```

```python
# 分子用変数
numer = 0.0

# 分母用変数
denom = 0.0

for i in range(len(df)):
    x = df.iloc[i]['平均時速 (km/h)']
    w = df.iloc[i]['距離']
    numer += w
    denom += w / x

numer / denom
```

```python
# 正解例

n = len(df)

# 分母
denom = 0.0

for i in range(n):
    x_j_prod = 1.0

    for j in range(n):
        x_j = df.iloc[j]['平均時速 (km/h)']
        x_j_prod *= x_j

    x_i = df.iloc[i]['平均時速 (km/h)']
    denom += x_j_prod / x_i

# 分子
numer = n

for j in range(n):
    x_j = df.iloc[j]['平均時速 (km/h)']
    numer *= x_j

numer / denom
```
