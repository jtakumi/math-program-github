```python
%matplotlib inline
import math  # 標準のmathモジュール

import numpy as np  # NumPy
import pandas as pd  # Pandas

from matplotlib import pylab as plt  # Matplotlib

# matplotlibで日本語表示する為の指定
import japanize_matplotlib
```

```python
baseball = pd.read_csv('../../9_data/プロ野球/プロ野球選手身長体重.csv')
baseball['BMI'] = baseball['体重'] / (baseball['身長'] / 100) ** 2
baseball.head(3)
```

```python
def sample_std(x):
    """
    標本標準偏差を求める
    :param x: 標本
    :return: 標本標準偏差
    """
    # 平均
    mu = np.mean(x)

    # 標本標準偏差
    sigma = 0.0

    # 要素数
    n = len(x)

    # 平均と要素の差を取って2乗する
    for i in range(n):
        sigma += (x[i] - mu) ** 2

    # 要素数 - 1で割る
    sigma /= (n - 1)

    # これが標本標準偏差の2乗になる
    # 最後に平方根を取れば標本標準偏差になる
    return np.sqrt(sigma)


# 野球のBMI標本標準偏差
baseball_std = sample_std(baseball['BMI'])

pd.DataFrame({'標本標準偏差': [baseball_std]}, index=['野球'])
```

```python
std = np.std(baseball['BMI'])
sample_std = np.std(baseball['BMI'], ddof=1)
pd.DataFrame({'標準偏差': [std], '標本標準偏差': [sample_std]}, index=['野球'])
```

```python
std = baseball['BMI'].std(ddof=0)
sample_std = baseball['BMI'].std()
pd.DataFrame({'標準偏差': [std], '標本標準偏差': [sample_std]}, index=['野球'])
```

```python
# 正解例

n = 100
x = np.random.choice(baseball['BMI'], n)
pd.DataFrame({'標準偏差': [x.std(ddof=0)], '標本標準偏差': [x.std(ddof=1)]}, index=['野球'])
```
