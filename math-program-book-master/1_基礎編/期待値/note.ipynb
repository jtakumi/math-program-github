```python
%matplotlib inline
import math  # 標準のmathモジュール

import numpy as np  # NumPy
import pandas as pd  # Pandas

from matplotlib import pylab as plt  # Matplotlib

# matplotlibで日本語表示する為の指定
import japanize_matplotlib
```

```python
# 合計本数
# (24ユニット * 2,000万本)
total_num = 24 * 20000000

df = pd.read_csv('../../9_data/宝くじ/「年末ジャンボ宝くじ」（第770回_全国自治宝くじ）の当せん金・本数.csv')
df = df.append(pd.Series(['外れ', 0, total_num - df['本数'].sum()], index=df.columns), ignore_index=True)
df
```

```python
r = 0

for m, p in zip(df['当せん金'], df['本数']):
    r += m * (p / total_num)

r
```

```python
np.average(df['当せん金'], weights=[p / total_num for p in df['本数']])
```
