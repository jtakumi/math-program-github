```python
df = pd.read_csv('../../9_data/経済/gdp.csv')
gdp = df['GDP実質成長率']
gdp = 1.0 + gdp / 100
pd.DataFrame(gdp.head())
```

```python
n = len(gdp)
la = 0.0

for i in range(n):
    la += np.log(gdp[i])

(np.exp(la / n) - 1) * 100
```
