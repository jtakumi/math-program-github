```python
from scipy.stats import zscore

points = [10, 20, 30, 40, 50, 60, 70, 80, 90]
standard_score = 10 * zscore(points) + 50

pd.DataFrame(standard_score, index=list(range(1, len(standard_score) + 1)), columns=['偏差値'])
```

```python
def standard_score(mu, sigma, x):
    """
    :param mu: 平均
    :param sigma: 標準偏差
    :param x: 偏差値を求めたい点数
    """
    return (10 * (x - mu)) / sigma + 50


mu = np.mean(points)
sigma = np.std(points)
data = list()

for x in points:
    data.append(standard_score(mu, sigma, x))

pd.DataFrame(data, index=list(range(1, len(data) + 1)), columns=['偏差値'])
```

```python
# 正解例

def standard_score(mu, sigma, x):
    """
    :param mu: 平均
    :param sigma: 標準偏差
    :param x: 偏差値を求めたい握力
    """
    return (10 * (x - mu)) / sigma + 50


standard_score(46.47, 7.51, 100)
```
