```python
%matplotlib inline
import math  # 標準のmathモジュール

import numpy as np  # NumPy
import pandas as pd  # Pandas

from matplotlib import pylab as plt  # Matplotlib

# matplotlibで日本語表示する為の指定
import japanize_matplotlib
```

```python
# グラフ生成用関数 (サイコロを500回振った際の平均値)

# 平均値
means = list()

# サイコロを振って出た目
numbers = np.array(list())

for n in np.random.choice(range(1, 7), size=500):
    numbers = np.append(numbers, n)
    means.append(numbers.mean())

plt.plot(means)
plt.axhline(y=3.5, color='r', linestyle='--')
```

```python
index = list()
data = {'平均値': list()}

# サイコロを振ったときに出た目の平均値を算出する
for i in range(8):
    n = 10 ** (i + 1)
    index.append(n)
    data['平均値'].append(np.random.choice(range(1, 7), size=n).mean())

pd.DataFrame(data, index=index)
```
