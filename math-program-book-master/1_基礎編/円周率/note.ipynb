```python
def get_pi(n_max):
    """
    piを求める関数
    n_max : 無限大にはできないので適当な大きな数字を入れる
    """
    pi = 0.0

    # 指定されたnの最大値までループする
    for n in range(n_max + 1):
        # 分子の部分 (-1のn乗)
        numer = (-1) ** n

        # 分母の部分 (2n + 1)
        denom = 2 * n + 1

        # 分子/分母を足していく
        pi += numer / denom

    # 上の計算でπ/4の値が出たので、最後に4倍すればπになる
    return pi * 4


get_pi(10000)
```
