```python
# グラフ用
from scipy.special import expit, logit

t = np.linspace(-4, 4, 100)
plt.plot(t, expit(t), label='シグモイド関数')
plt.plot(t, logit(t), label='ロジット関数')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('図2. シグモイド関数')
```

```python
from scipy.special import expit


def logit(x):
    """
    ロジット
    :param x: x
    :return: ロジット
    """
    return np.log(x / (1 - x))


data = list()
points = [4, 8, 15, 16, 23, 42]

for a in points:
    data.append(logit(expit(a)))

pd.DataFrame(data, index=points, columns=['標準シグモイド関数->ロジット'])
```

```python
from decimal import Decimal


def logit(x):
    """
    ロジット
    :param x: x
    :return: ロジット
    """
    return math.log(x / (1 - x))


def sigmoid(x):
    """
    標準シグモイド関数
    :param x: x
    :return: 標準シグモイド関数
    """
    return 1 / (1 + np.exp(-x))


a = 42
pd.DataFrame(logit(sigmoid(Decimal(a))), index=[a], columns=['標準シグモイド関数->ロジット'])
```
