```python
def std_sigmoid(x):
    """
    標準シグモイド関数
    :param x: x
    :return: 標準シグモイド関数
    """
    return 1 / (1 + np.exp(-x))


t = np.linspace(-6, 6, 100)
plt.plot(t, std_sigmoid(t))
plt.title('グラフ1. 標準シグモイド関数')
```

```python
def sigmoid(x, a):
    """
    シグモイド関数
    :param x: x
    :param a: ゲイン
    :return: シグモイド関数
    """
    return 1 / (1 + np.exp(-a * x))


t = np.linspace(-5, 5, 100)
plt.figure(figsize=(10, 5))

# a=1〜4までのカーブをplot
for a in range(1, 5):
    plt.plot(t, sigmoid(t, a), label='a=%d' % a)

# グラフにlegendを表示
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# 正解例

def std_sigmoid(x):
    """
    標準シグモイド関数
    :param x: x
    :return: 標準シグモイド関数
    """
    return 1 / (1 + np.exp(-x))


def std_sigmoid2(x):
    """
    標準シグモイド関数 (tanh使用)
    :param x: x
    :return: 標準シグモイド関数 (tanh使用)
    """
    return (np.tanh(x / 2) + 1) / 2


data = list()
points = np.linspace(-1, 1, 10)

for t in points:
    data.append([std_sigmoid(t), std_sigmoid2(t)])

pd.DataFrame(data, index=points, columns=['標準シグモイド関数', '標準シグモイド関数 (tanh使用)'])
```
