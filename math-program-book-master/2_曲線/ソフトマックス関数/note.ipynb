```python
%matplotlib inline
import math  # 標準のmathモジュール

import numpy as np  # NumPy
import pandas as pd  # Pandas

from matplotlib import pylab as plt  # Matplotlib

# matplotlibで日本語表示する為の指定
import japanize_matplotlib
```

```python
def softmax(x, j):
    """
    ソフトマックス関数
    :param x: 入力
    :param j: インデックス
    :return: ソフトマックス関数
    """
    return np.exp(x[j]) / np.sum(np.exp(x))


# 1〜10の値が入ったndarray
x = np.arange(1, 11)

result = list()

# 各要素に対してsoftmax関数を実行
for j in range(10):
    result.append(softmax(x, j))

_, ax1 = plt.subplots()
ax1.plot(result)
ax1.twinx().plot(x, color='r')
```

```python
from sklearn.utils.extmath import softmax

# 1〜10の値が入ったndarray
x = np.arange(1.0, 11, 1)

# 各要素に対してsoftmax関数を実行
result = softmax([x])

_, ax1 = plt.subplots()
ax1.plot(result[0])
ax1.twinx().plot(x, color='r')
```

```python
# 正解例

from scipy.special import expit


def sigmoid_softmax(x):
    """
    標準シグモイド関数 (ソフトマックス関数使用)
    :param x: x
    :return: 標準シグモイド関数 (ソフトマックス関数使用)
    """
    x = [x, 0]
    return np.exp(x[0]) / np.sum(np.exp(x))


t = np.linspace(-6, 6, 10)
result = list()

for j in t:
    result.append([expit(j), sigmoid_softmax(j)])

pd.DataFrame(result, index=t, columns=['標準シグモイド関数', '標準シグモイド関数 (ソフトマックス関数使用)'])
```
