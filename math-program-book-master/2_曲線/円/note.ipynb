```python
# 100個のポイントを作成
t = np.linspace(-np.pi, np.pi, 100)

# x軸とy軸の長さが等しくなるように設定
plt.figure(figsize=(5, 5))

# x軸はcos、y軸はsinで描画する
plt.plot(np.cos(t), np.sin(t))
```

```python
# 100個のポイントを作成
t = np.linspace(-np.pi, np.pi, 100)

# x軸とy軸の長さが等しくなるように設定
plt.figure(figsize=(5, 5))

# x軸はcos、y軸はsinの1/2で描画する
plt.plot(np.cos(t), 1 / 2 * np.sin(t))

plt.ylim((-1.0, 1.0))
```

```python
# 100個のポイントを作成
t = np.linspace(-np.pi, np.pi, 100)

# x軸とy軸の長さが等しくなるように設定
plt.figure(figsize=(5, 5))

# 元の楕円のx軸
x = np.cos(t)

# 元の楕円のy軸
y = 1 / 2 * np.sin(t)

# 反時計回りに45度傾ける
r_rad = np.pi / 4

plt.plot(x * np.cos(r_rad) - y * np.sin(r_rad), x * np.sin(r_rad) + y * np.cos(r_rad))
```

```python
n_points = 500

# 円10周分のポイントを用意
t = np.linspace(-np.pi * 5, np.pi * 15, n_points)

# 渦巻きを作る為の倍率
r = np.linspace(0.0, 1.0, n_points)

# x軸とy軸の長さが等しくなるように設定
plt.figure(figsize=(5, 5))

plt.plot(np.cos(t) * r, np.sin(t) * r)
```

```python
# 正解例

t = np.linspace(-np.pi * 5, np.pi * 15, 500)
plt.figure(figsize=(5, 5))
plt.plot(np.cos(t) ** 3, np.sin(t) ** 3)
```
