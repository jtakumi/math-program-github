```python
%matplotlib inline
import math  # 標準のmathモジュール

import numpy as np  # NumPy
import pandas as pd  # Pandas

from matplotlib import pylab as plt  # Matplotlib

# matplotlibで日本語表示する為の指定
import japanize_matplotlib
```

```python
# グラフ生成用関数
from scipy.special import erf

t = np.linspace(-np.pi, np.pi, 1000)
plt.plot(t, erf(t))
plt.title('グラフ1 : 誤差関数')
```

```python
# グラフ生成用関数
from scipy.special import erf
from scipy.special import expit

t = np.linspace(-5, 5, 1000)
plt.plot(t, erf(t), label='誤差関数')
plt.plot(t, expit(t) * 2 - 1, label='シグモイド関数 × 2 - 1')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('グラフ2 : 誤差関数とシグモイド関数 × 2 - 1')
```

```python
def f(t):
    """
    積分する対象になる関数
    :param t: t
    :return: 積分する対象になる関数
    """
    return np.e ** (-t ** 2)
```

```python
from scipy import integrate

integrate.quad(f, 0, 1)[0] * (2 / np.sqrt(np.pi))
```

```python
# 正解例

erf = 0.0

for n in range(1000):
    erf += ((-1) ** n * 1 ** (2 * n + 1)) / (math.factorial(n) * (2 * n + 1))

(2 / np.sqrt(np.pi)) * erf
```
