```python
%matplotlib inline
import math  # 標準のmathモジュール

import numpy as np  # NumPy
import pandas as pd  # Pandas

from matplotlib import pylab as plt  # Matplotlib

# matplotlibで日本語表示する為の指定
import japanize_matplotlib
```

```python
# sinとcos
point_range = 5 * np.pi
points = np.linspace(-point_range, point_range, 100)
plt.plot(np.sin(points))
plt.plot(np.cos(points))
plt.title('sin, cos')
```

```python
def sin(z, n_lim=100):
    """
    sineを求める
    :param z: 角度
    :param n_lim: nの上限
    :return: sine
    """
    _sin = 0

    for n in range(n_lim):
        m = 2 * n + 1
        _sin += (((-1) ** n) / math.factorial(m)) * (z ** m)

    return _sin


rads = [0, 0.5, 1, 1.5]
pd.DataFrame([[sin(r), np.sin(r)] for r in rads], index=rads, columns=['自作関数', 'NumPy'])
```

```python
# オーバーフローを起こす例
sin(2, 1000)
```

```python
# 正解例

def cos(z, n_lim=100):
    """
    cosineを求める
    :param z: 角度
    :param n_lim: nの上限
    :return: cosine
    """
    _cos = 0

    for n in range(n_lim):
        m = 2 * n
        _cos += (((-1) ** n) / math.factorial(m)) * (z ** m)

    return _cos


rads = [0, 0.5]
pd.DataFrame([[cos(r), np.cos(r)] for r in rads], index=rads, columns=['自作関数', 'NumPy'])
```

```python
# 正解例

def tan(z, n_lim=100):
    """
    tangentを求める
    :param z: 角度
    :param n_lim: nの上限
    :return: tangent
    """
    return sin(z, n_lim) / cos(z, n_lim)


rads = [1, 1.5]
pd.DataFrame([[tan(r), np.tan(r)] for r in rads], index=rads, columns=['自作関数', 'NumPy'])
```
